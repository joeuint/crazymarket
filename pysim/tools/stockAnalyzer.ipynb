```python
import matplotlib.pyplot as plt
```

# Read File

```python
def load_stock_price_hist(path):
    with open(path, 'r') as f:
        return [float(line.strip()) for line in f if line.strip()]

numbers = load_stock_price_hist("../ACME_price.txt")

print(numbers[:10])
```

# Plot Price

```python
plt.plot(numbers[:200])
plt.title("Stock Price History (First 200 Entries)")
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()
```
