```python
import matplotlib.pyplot as plt
```

# Read File

```python
def load_stock_price_hist(path):
    with open(path, 'r') as f:
        return [float(line.strip()) for line in f if line.strip()]

stock1 = load_stock_price_hist("../Picon Inc._price.txt")
stock2 = load_stock_price_hist("../Banana Corp._price.txt")

print(stock1[:10])
```

# Plot Price

```python
plt.plot(stock1, label="ACME")
plt.plot(stock2, label="FOOBAR")
plt.title("Stock Price History")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```
